## Preparing the dataset

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
# Load results data
file = 'ExpedientesPublicados2020_200906074133.csv'
df = pd.read_csv(file)
print('Dataframe shape:')

# Trimestre - esto toma mucho tiempo asi que solo lo hago una vez y modifico el archivo original yolo
# df['Trimestre'] = df['Publicación del anuncio'].map(lambda x: pd.Period(x,'Q'))
# df.to_csv('ExpedientesPublicados2020_200906074133.csv', index = False)

display(df.shape)
df.head()
```

### Remove irrelevant attributes

```python
columns_to_drop = ['Vigencia del anuncio', 'Fecha de creación', 'Fecha de última modificacion', 'Código del expediente', 'Número del procedimiento', 'Dirección del anuncio', 'Clave de la UC', 'Clave COG']
df = df.drop(columns_to_drop, axis=1)
columns_to_drop = ['Forma del procedimiento', 'Artículo de excepción', 'REFERENCIA_EXPEDIENTE', 'Título del expediente', 'Nombre de la UC', 'Operador']
# columns_to_drop = ['Entidad federativa', 'Correo electrónico', 'Plantilla del expediente']
df = df.drop(columns_to_drop, axis=1)
print('Dataframe shape:')
display(df.shape)
df.head()
```

### Add New Attributes

```python
# Add column for Prodecimiento
df['Procedimiento_isNacional'] = np.where(df['Carácter del procedimiento']== 'Nacional', 1, 0)
df['Procedimiento_isInternacional'] = np.where(df['Carácter del procedimiento']== 'Interacional', 1, 0)
df['Procedimiento_isIntTLC'] = np.where(df['Carácter del procedimiento']== 'Internacional bajo TLC', 1, 0)

# Add columns for Tipo de Contratacion
df['Contratacion_isAdquisiciones'] = np.where(df['Tipo de contratación']== 'Adquisiciones', 1, 0)
df['Contratacion_isArriendamientos'] = np.where(df['Tipo de contratación']== 'Arriendamientos', 1, 0)
df['Contratacion_isObraPublica'] = np.where(df['Tipo de contratación']== 'Obra Pública', 1, 0)
df['Contratacion_isServicios'] = np.where(df['Tipo de contratación'].str.contains('Servicios', na=False, regex=True), 1, 0)

# Add column for the occurrances of each State
df['Entidad_count'] = df.groupby('Entidad federativa')['Entidad federativa'].transform('count')

# Add column for whether the buyer used an official gob email address
df['Correo_isGob'] = np.where(df['Correo electrónico'].str.contains('.gob.mx', na=False, regex=False), 1, 0)

# Add columns for Plantilla: Licitacion, adjudicacion, invitacion
df['Plantilla_isLicitacion'] = np.where(df['Plantilla del expediente'].str.contains('Licitación', na=False, regex=False), 1, 0)
df['Plantilla_isAdjudicacion'] = np.where(df['Plantilla del expediente'].str.contains('Adjudicación', na=False, regex=False), 1, 0)
df['Plantilla_isInvitacion'] = np.where(df['Plantilla del expediente'].str.contains('Invitación', na=False, regex=False), 1, 0)

# Add columns for each Trimestre on Publicacion del anuncio
df['Publicacion_isTrimestre1'] = np.where(df['Trimestre']== '2020Q1', 1, 0)
df['Publicacion_isTrimestre2'] = np.where(df['Trimestre']== '2020Q2', 1, 0)
df['Publicacion_isTrimestre3'] = np.where(df['Trimestre']== '2020Q3', 1, 0)
df['Publicacion_isTrimestre4'] = np.where(df['Trimestre']== '2020Q4', 1, 0)

# Add NLP columns
df['Descripcion_isSalud'] = np.where(df['Descripción del Anuncio'].str.contains('medicamento', na=False, regex=False), 1, 0)
df['Descripcion_isCovid'] = np.where(df['Descripción del Anuncio'].str.contains('covid', na=False, regex=False), 1, 0)
df['Descripcion_isProfesional'] = np.where(df['Descripción del Anuncio'].str.contains('profesional', na=False, regex=False), 1, 0)
df['Descripcion_isMateri'] = np.where(df['Descripción del Anuncio'].str.contains('materi', na=False, regex=False), 1, 0)
df['Descripcion_isMantenimiento'] = np.where(df['Descripción del Anuncio'].str.contains('mantenimiento', na=False, regex=False), 1, 0)
df['Descripcion_isPrograma'] = np.where(df['Descripción del Anuncio'].str.contains('programa', na=False, regex=False), 1, 0)
df['Descripcion_isSuministro'] = np.where(df['Descripción del Anuncio'].str.contains('suministro', na=False, regex=False), 1, 0)

# Drop the extra columns
columns_to_drop = ['Entidad federativa', 'Correo electrónico', 'Plantilla del expediente', 'Carácter del procedimiento', 'Tipo de contratación',  'Publicación del anuncio', 'Trimestre', 'Descripción del Anuncio']
df = df.drop(columns_to_drop, axis=1)

# # Display dataset
print('Dataframe shape:')
display(df.shape)
display(df)

df.to_csv('dataset_v3.csv', index = False)
```
